(b) iii. Select the ﬁrst 70 rows of Class 0 and the ﬁrst 140 rows of Class 1 as the training set and the rest of the data as the test set.
1. training set
Class 0(NO): row[210:280], Class 1(AB): row[0:140]
2. test set
Class 0(NO): row[280:],    Class 1(AB): row[140:210]

```python
train1 = df2.iloc[0:140, :] # 140 rows
train2 = df2.iloc[210:280, :] # 70 rows

train = pd.concat([train1, train2]) # 210 rows

test1 = df2.iloc[140:210, :] # 70 rows
test2 = df2.iloc[280:, :] # 30 rows

test = pd.concat([test1, test2]) # 100 rows
```

(c) Classiﬁcation using KNN on Vertebral Column Data Set
    
    i. Write code for k-nearest neighbors with Euclidean metric (or use a software package). 

```python
from sklearn.neighbors import KNeighborsClassifier

train_X = train.iloc[:, :6]     # train data
train_y = train.iloc[:, 6]      # class 

def KNN_dtm (X, y, k, tX):    # tX: test data set
    neigh = KNeighborsClassifier(n_neighbors=k)  # defaulted p = 2, defaulted metric = ‘minkowski’: Euclidean_distance    
    neigh.fit(X, y)
    pred = neigh.predict(tX)
    
    return pred
```

c. ii. Test all the data in the test database with k nearest neighbors. Take decisions by majority polling.

```python
test_X = test.iloc[:, :6]   # test data
test_y = test.iloc[:, 6]    # original class of test data

for i in range(70):
    k = 3*i+1    # k = 1,4,7,..,208
    print(KNN_dtm (train_X, train_y, k, test_X)) # defaluted weights=’uniform’: majority polling
```

(c) ii. Plot train and test errors in terms of k for k ∈{208,205,...,7,4,1} (in reverse order). You are welcome to use smaller increments of k. 

```python
# train_X = train.iloc[:, :6]     # train data
# train_y = train.iloc[:, 6]      # class 
# test_X  = test.iloc[:, :6]      # test data
# test_y  = test.iloc[:, 6]       # original class of test data


from sklearn.metrics import accuracy_score  # to get error rate

# training error rate: trainER

dic = dict ()
for i in range(70):
    k = 3*i+1    # k = 1,4,7,..,208
    predtr = KNN_dtm (train_X, train_y, k, train_X)   # testing train data
    
    # calculating error rate
    ac = accuracy_score(train.iloc[:, 6], predtr)     # comparing true class and predicted class
    dic[i] = k, 1-ac
    
trainER = pd.DataFrame(dic).T

# print(trainER)

# test error rate: testER

dic = dict ()
for i in range(70):
    k = 3*i+1    # k = 1,4,7,..,208
    predte = KNN_dtm (train_X, train_y, k, test_X)   # testing test data
    
    # calculating error rate
    ac = accuracy_score(test.iloc[:, 6], predte)     # comparing true class and predicted class
    dic[i] = k, 1-ac
          
testER = pd.DataFrame(dic).T

# print(testER)
```

(c) ii. Plot train and test errors in terms of k for k ∈{208,205,...,7,4,1} (in reverse order)

```python
plt.plot(trainER.iloc[:,0], trainER.iloc[:,1])  # train error rate
plt.plot(testER.iloc[:,0], testER.iloc[:,1])    # test error rate
plt.gca().invert_xaxis()

plt.title('Error Rate per k')
plt.xlabel('k')
plt.ylabel('Error Rate')
plt.legend(['train error rate', 'test error rate'])
plt.show()
```

(c) ii. Which k∗ is the most suitable k among those values? 

```python
s_testER = testER.sort_values(by=[1, 0])  # sorting testER(k, test error) by test error and k

# select best k with the least test error rate 
# If many k's give the same error, select the least k except for k=1 to avoid overfitting and biased

if s_testER.iloc[0, 0] == 1 and round(s_testER.iloc[0, 1], 3) == round(s_testER.iloc[1, 1], 3):
    bst_k = int(s_testER.iloc[1, 0])
    bst_er = s_testER.iloc[1, 1]
else:
    bst_k = int(s_testER.iloc[0, 0])
    bst_er = s_testER.iloc[0, 1]

print("k*=", bst_k, "with best test error:", round(bst_er, 3))    
print(s_testER)  # k, test error
```

```python
def prec (TP, FP):      # TP: 'AB' people predicted as 'AB'
    PPV = TP /(TP+FP)   # FP: 'NO' people predicted as 'AB'
    return PPV

PPV = prec (cm[0][0], cm[1][0]) # precision
print(PPV)  
```

(c) iii. a subset of the training set N ∈ {10,20,30,...,210}
for each N, select your training set by choosing the ﬁrst  [N/3] rows of Class 0 and the ﬁrst N −[N/3] rows of Class 1 in the training set you creatd in 1(b)iii. 

```python
# train1 = df2.iloc[0:140, :] # 140 rows
# train2 = df2.iloc[210:280, :] # 70 rows

# train = pd.concat([train1, train2]) # 210 rows

# test1 = df2.iloc[140:210, :] # 70 rows
# test2 = df2.iloc[280:, :] # 30 rows

# test = pd.concat([test1, test2]) # 100 rows

# train_X = train.iloc[:, :6]     # train data
# train_y = train.iloc[:, 6]      # class 
# test_X  = test.iloc[:, :6]      # test data
# test_y  = test.iloc[:, 6]       # original class of test data

# function for making new train sets

import math  # to use floor function

def mkntrain (N):
    n = math.floor(N/3)
    
    ntrain1 = df2.iloc[0:N-n, :]     # the first N-n rows of Class 1 (AB)
    ntrain2 = df2.iloc[210:210+n, :] # the first n rows of Class 0 (NO)

    ntrain = pd.concat([ntrain1, ntrain2]) # N rows of trains
    
    return ntrain
```

(c) iii. the optimal k is selected from {1,6,11,...,196}. 

```python
# function for KNN
# def KNN_dtm (X, y, k, tX):    # tX: test data set
#    neigh = KNeighborsClassifier(n_neighbors=k)  # defaulted p = 2, defaulted metric = ‘minkowski’: Euclidean_distance    
#    neigh.fit(X, y)
#    pred = neigh.predict(tX)
    
#    return pred

dt = dict()

for n in range(21):    
    N = 10*(n+1)      # N ∈ {10,20,30,...,210}
    
    ntrain = mkntrain (N)
    tr_X = ntrain.iloc[:, :6]     # subset train data
    tr_y = ntrain.iloc[:, 6]      # class

    dic = dict ()
    
    testerr = None
    s_testerr = None
    
    for i in range(int(N/5)):  
        k = 5*i+1    # k = 1,6,...
        predte = KNN_dtm (tr_X, tr_y, k, test_X)   # testing all test data for N, k

        # calculating error rate
        ac = accuracy_score(test.iloc[:, 6], predte)     # comparing true class and predicted class
        dic[i] = k, 1-ac
          
    testerr = pd.DataFrame(dic).T   # test error rate for N & all k
    
    # the best test error rate, and the best k for N
    
    s_testerr = testerr.sort_values(by=[1, 0])  # sorting testerr(k, test error) by test error and k

# select best k with the least test error rate 
# If many k's give the same error, 
# select the least k except for k=1 to avoid overfitting and biased 

    if s_testerr.iloc[0, 0] == 1 and round(s_testerr.iloc[0, 1],3) == round(s_testerr.iloc[1, 1], 3):
        bst_k = int(s_testerr.iloc[1, 0])
        bst_er = s_testerr.iloc[1, 1]
    else:
        bst_k = int(s_testerr.iloc[0, 0])
        bst_er = s_testerr.iloc[0, 1]
     
    dt[n]= N, bst_k, bst_er # N, opt_k, best test error rate
    
nbo = pd.DataFrame(dt).T

print(nbo) # N, opt_k, best error rate 
```

(d) i. B. Minkowski Distance with log10(p) ∈ {0.1,0.2,0.3,...,1}. In this case, use the k∗ you found for the Manhattan distance in 1(d)iA. What is the best log10(p)? 

```python
# k∗ = 11
# p = 10**0.1, 10**0.2,...10**1

def KNN_dtm3 (X, y, k, tX, p_in):    # tX: test data set
    neigh = KNeighborsClassifier(n_neighbors=k, p=10**(0.1*p_in))  # defaulted metric = ‘minkowski’    
    neigh.fit(X, y)
    pred = neigh.predict(tX)
    
    return pred

dic = dict ()

for i in range(10):
    p_in = i+1    #  p = 10**0.1, 10**0.2,...10**1 (p=10**(0.1*p_in)
    predte = KNN_dtm3 (train_X, train_y, 11, test_X, p_in) # testing test data with all train data
    
    # calculating error rate
    ac = accuracy_score(test.iloc[:, 6], predte)     # comparing true class and predicted class
    dic[i] = 0.1*p_in, 1-ac    # test error rate for log10(p) = log10(10**(0.1*p_in)) = 0.1*p_in 
          
testER3 = pd.DataFrame(dic).T

# sorting testER(log10(p), test error) by test error and log10(p)

s_testER3 = testER3.sort_values(by=[1, 0])  

print(s_testER3)  # log10(p), test error rate
```

ii. Mahalanobis Distance.

(Mahalanobis Distance requires inverting the covariance matrix of the data. When the covariance matrix is singular or ill-conditioned, the data live in a linear subspace of the feature space. In this case, the features have to be transformed into a reduced feature set in the linear subspace, which is equivalent to using a pseudoinverse instead of an inverse.)

```python
cov=np.cov(train_X, rowvar=False)  # cov = covariance matrix of the train data

def KNN_dtmd (X, y, k, tX):    # tX: test data set
    neigh = KNeighborsClassifier(n_neighbors=k, metric = 'mahalanobis', metric_params={'V': cov})
    neigh.fit(X, y)
    pred = neigh.predict(tX)
    
    return pred

dic = dict ()
for i in range(40):
    k = 5*i+1    # k = 1,6,11,...,196
    predte = KNN_dtmd (train_X, train_y, k, test_X)  # testing test data with all train data

    # calculating error rate
    ac = accuracy_score(test.iloc[:, 6], predte)     # comparing true class and predicted class
    dic[i] = k, 1-ac
          
testERd = pd.DataFrame(dic).T

s_testERd = testERd.sort_values(by=[1, 0])  # sorting testER(k, test error) by test error and k
    
# select best k with the least test error rate 
# If many k's give the same error, 
# select the least k except for k=1 to avoid overfitting and biased  

if s_testERd.iloc[0, 0] == 1 and round(s_testERd.iloc[0, 1], 3) == round(s_testERd.iloc[1, 1], 3):
    bst_k = int(s_testERd.iloc[1, 0])
    bst_er = s_testERd.iloc[1, 1]
else:
    bst_k = int(s_testERd.iloc[0, 0])
    bst_er = s_testERd.iloc[0, 1]

print("Mahalanobis Distance, k*=", bst_k, "with best test error:", round(bst_er, 3))    
print(s_testERd)  # k, test error
```
